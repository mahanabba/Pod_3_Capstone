```python
tmdbDF = pd.read_csv("zippedData/tmdb.movies.csv")
tmdbDF.head()
```

```python
tnDF = pd.read_csv("zippedData/tn.movie_budgets.csv")
tnDF.head()
```

```python
import sqlite3
conn = sqlite3.connect('zippedData/im.db')
```

```python
movie_basicsDF = pd.read_sql("select * from movie_basics", con=conn)
movie_basicsDF.head()
```

```python
movie_ratingsDF = pd.read_sql("select * from movie_ratings", con=conn)
movie_ratingsDF.head()
```

```python
pd.read_sql("select * from movie_akas", con=conn)
```

```python

tnDF["production_budget"] = tnDF["production_budget"].str.replace("$", "")
tnDF["domestic_gross"] = tnDF["domestic_gross"].str.replace("$", "")
tnDF["worldwide_gross"] = tnDF["worldwide_gross"].str.replace("$", "")



tnDF["production_budget"] = tnDF["production_budget"].str.replace(",", "")
tnDF["domestic_gross"] = tnDF["domestic_gross"].str.replace(",", "")
tnDF["worldwide_gross"] = tnDF["worldwide_gross"].str.replace(",", "")



tnDF["production_budget"] = tnDF["production_budget"].astype(int)
tnDF["domestic_gross"] = tnDF["domestic_gross"].astype(int)
tnDF["worldwide_gross"] = tnDF["worldwide_gross"].astype(float)
```

```python

tnDF["domestic_profit"] = tnDF["domestic_gross"] - tnDF["production_budget"]

```

```python
budget_ratings_merge = pd.merge(tnDF, tmdbDF, left_on='movie', right_on="title")
budget_ratings_merge.head()
```

```python
import matplotlib.pyplot as plt
budget_ratings_merge.plot(x="vote_average", y="domestic_profit", kind="scatter")
```

```python
budget_ratings_merge.plot(x="vote_average", y="worldwide_gross", kind="scatter")
```

```python
budget_ratings_merge.plot(x="vote_average", y="domestic_gross", kind="scatter")
```
